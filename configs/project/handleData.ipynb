```python
import matplotlib.pyplot as plt

file_path = "/home/infres/rcarvalho-23/gem5/m5out/data.txt"

ticks = []
instrInROB = []
instrFin = []
instrExec = []
instrWait = []

x_start = 0
x_end = 1e3

with open(file_path, "r") as file:
    next(file)
    for line in file:
        columns = line.strip().split("\t")
        ticks.append(int(float(columns[0])/1000))
        instrInROB.append(int(columns[1]))
        instrFin.append(int(columns[2]))
        instrExec.append(int(columns[3]))
        instrWait.append(int(columns[4]))

fig = plt.figure(figsize=(8, 6))

plot1 = fig.add_subplot(221)
plot1.plot(ticks, instrInROB)
plot1.set_xlabel("Clock Cycles")
plot1.set_ylabel("Instr in ROB")

plot2 = fig.add_subplot(222)
plot2.plot(ticks, instrFin)
plot2.set_xlabel("Clock Cycles")
plot2.set_ylabel("Instr to Commit")

plot3 = fig.add_subplot(223)
plot3.plot(ticks, instrExec)
plot3.set_xlabel("Clock Cycles")
plot3.set_ylabel("Instr to Exec")

plot4 = fig.add_subplot(224)
plot4.plot(ticks, instrWait)
plot4.set_xlabel("Clock Cycles")
plot4.set_ylabel("Instr Wait")

plt.tight_layout()
plt.show()

fig2, ax = plt.subplots(figsize=(8, 6))

ax.plot(ticks, instrInROB, label="Instr in ROB", color="blue")
ax.plot(ticks, instrFin, label="Instr to Commit", color="red")
ax.plot(ticks, instrExec, label="Instr to Exec", color="green")
ax.plot(ticks, instrWait, label="Instr Wait", color="purple")

ax.set_xlabel("Clock Cycles")
ax.set_ylabel("Number of Instructions")
ax.set_title("Instruction Over Time")
ax.legend()
plt.grid(True)
plt.show()

fig3, ax_zoom = plt.subplots(figsize=(8, 6))
ax_zoom.plot(ticks, instrInROB, label="Instr in ROB", color="blue")
ax_zoom.plot(ticks, instrFin, label="Instr to Commit", color="red")
ax_zoom.plot(ticks, instrExec, label="Instr to Exec", color="green")
ax_zoom.plot(ticks, instrWait, label="Instr Wait", color="purple")

ax_zoom.set_xlabel("Clock Cycles")
ax_zoom.set_ylabel("Number of Instructions")
ax_zoom.set_title("Zoomed View of Instruction Metrics")
ax_zoom.legend()
plt.grid(True)
ax_zoom.set_xlim(x_start, x_end)
plt.show()

fig4 = plt.figure(figsize=(8, 6))

plot1_zoom = fig4.add_subplot(221)
plot1_zoom.plot(ticks, instrInROB)
plot1_zoom.set_xlabel("Clock Cycles")
plot1_zoom.set_ylabel("Instr in ROB")
plot1_zoom.set_xlim(x_start, x_end)

plot2_zoom = fig4.add_subplot(222)
plot2_zoom.plot(ticks, instrFin)
plot2_zoom.set_xlabel("Clock Cycles")
plot2_zoom.set_ylabel("Instr to Commit")
plot2_zoom.set_xlim(x_start, x_end)

plot3_zoom = fig4.add_subplot(223)
plot3_zoom.plot(ticks, instrExec)
plot3_zoom.set_xlabel("Clock Cycles")
plot3_zoom.set_ylabel("Instr to Exec")
plot3_zoom.set_xlim(x_start, x_end)

plot4_zoom = fig4.add_subplot(224)
plot4_zoom.plot(ticks, instrWait)
plot4_zoom.set_xlabel("Clock Cycles")
plot4_zoom.set_ylabel("Instr Wait")
plot4_zoom.set_xlim(x_start, x_end)

plt.tight_layout()
plt.show()
```
